Visualisation

Hour Studied vs Exam_Note

```python
plt.bar(df['Hours_Studied'], df['Exam_Score'])
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Exam Score vs Hours Studied')

# Display the plot
plt.show()
```

Scores Comparison

```python
plt.plot(df['Previous_Scores'], label='Previous Scores')
plt.plot(df['Exam_Score'], label='Exam Score')
plt.title('Scores Comparison')
plt.legend()
plt.show()
```

Sleep Hours Vs Note Exams

```python
sns.violinplot(x='Sleep_Hours', y='Exam_Score', data=df)

# Adding labels and title
plt.xlabel('Sleep Hours')
plt.ylabel('Exam Score')
plt.title('Sleep Hours vs Exam Score Distribution')

# Display the plot
plt.show()
```

Gender vs Exam Score

```python
sns.barplot(x='Gender', y='Exam_Score', data=df)
plt.title('Gender vs Exam Score')
plt.show()
```

Répartition des types d'école

```python
school_type_counts = df['School_Type'].value_counts()  
colors = ['red','#66b3ff']  
plt.pie(school_type_counts, labels=school_type_counts.index, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Répartition des types d\'école')
plt.axis('equal')
plt.show()
```

Teacher Quality vs Exam Score

```python
sns.boxplot(x='Exam_Score', y='Teacher_Quality', data=df)
plt.xlabel('Exam Score')
plt.ylabel('Teacher Quality')
plt.title('Boxplot of Teacher Quality vs Exam Score')
plt.show()
```
